```python
good = set()

good_images_path = 'cleanup.json'
with open(good_images_path) as f:
    good_images = json.load(f)
    
for name in good_images:
    if good_images[name] == '' or good_images[name] == '1':
        good.add(name)
good_list = list(good)
print(good_list[0:10])
print(len(good_list))
```

```python
import random

random.shuffle(good_list)

print(good_list[0:10])
```

```python
test = good_list[0:170]
val = good_list[170: 340]
train = good_list[340:]
print(len(train))
```

```python
test_json = {}
val_json = {}
train_json = {}

for k in ['categories', 'licenses', 'info']:
    test_json[k] = data[k]
    val_json[k] = data[k]
    train_json[k] = data[k]

test_json['images'] = []
val_json['images'] = []
train_json['images'] = []
test_json['annotations'] = []
val_json['annotations'] = []
train_json['annotations'] = []

test_ids = set()
val_ids = set()
train_ids = set()

print(test_json)
```

```python
for image in data['images']:
    if image['file_name'] in test:
        test_json['images'].append(image)
        test_ids.add(image['id'])
    elif image['file_name'] in val:
        val_json['images'].append(image)
        val_ids.add(image['id'])
    elif image['file_name'] in train:
        train_json['images'].append(image)
        train_ids.add(image['id'])
        
```

```python
for annotation in data['annotations']:
    if annotation['image_id'] in test_ids:
        test_json['annotations'].append(annotation)
    elif annotation['image_id'] in val_ids:
        val_json['annotations'].append(annotation)
    elif annotation['image_id'] in train_ids:
        train_json['annotations'].append(annotation)
    
```

```python
ids = set()
for im in data['images']:
    ids.add(im['id'])
print(len(ids))
```

```python
pids = set()
for ann in data['annotations']:
    pids.add(ann['image_id'])
print(len(pids))
```

```python
json_object = json.dumps(test_json)
 
# Writing to sample.json
with open("test_clean_all_bbox.json", "w") as outfile:
    outfile.write(json_object)
```
