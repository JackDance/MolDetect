```python
import os
import json
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
```

## BBox

```python
from rxnscribe.data import ImageData

image_base = '../data/detect/images/'
data_path = '../data/detect/splits/annotations/output.json'

with open(data_path) as f:
    data = json.load(f)
    
for idx, image_data in enumerate(data['images']):
    path = image_base + image_data['file_name']
    img = cv2.imread(path)
    d = ImageData(image_data)
    fig, ax = plt.subplots(figsize=(10, 10))
    d.draw_gold(ax, img)
    # You can save the img to a file here.    
    plt.show()
    if idx == 3:
        break
```

## Predictions

```python
pred_path = '../output/pix2seq_bbox_rand32_unique_bbox_600_4e-4_composite_heuristic/prediction_newtest.json'
image_base = '../data/detect/pdffigures/images/'

with open(pred_path) as f:
    predictions = json.load(f)

for i, file_name in enumerate(predictions['file_name']):
    image_file = image_base + file_name
    img = cv2.imread(image_file)
    d = ImageData(predictions=predictions['bbox'][i], image_file=image_file)
    ig, ax = plt.subplots(figsize=(10, 10))
    d.draw_prediction(ax, img)
    plt.show()
    if i == 50:
        break
```
