```python
import os
from predict_bbox import get_args, predict_image, BboxTokenizer, ReactionExtractorPix2Seq

args = get_args(notebook=True)
args.save_path = "/Mounts/rbg-storage1/users/yujieq/reaction/output/pix2seq_bbox_rand/"
args.image_path = "/Mounts/rbg-storage1/users/yujieq/reaction/data/detect/images/acs.joc.5b00301-Figure-c8.png"

tokenizer = BboxTokenizer(input_size=2000, sep_xy=False, pix2seq=True, rand_target=False)
model = ReactionExtractorPix2Seq.load_from_checkpoint(
    os.path.join(args.save_path, 'checkpoints/best.ckpt'), strict=False, args=args, tokenizer=tokenizer)
model.eval()
model.cuda()

prediction = predict_image(model, args.image_path)
```

```python
import cv2
import matplotlib.pyplot as plt
from reaction.data import BBox

img = cv2.imread(args.image_path)
h, w, _ = img.shape
boxes = [BBox(box, w, h, xyxy=True, normalized=True) for box in prediction]

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(img)
for box in boxes:
    box.draw(ax)
```
